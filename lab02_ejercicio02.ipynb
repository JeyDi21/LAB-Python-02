```python
import pandas as pd
# Data de prueba
numeros = [5, 8, 12, 15, 3, 20]
temperaturas_celsius = [0, 20, 30, 100]
palabras = ["manzana", "sol", "programacion", "hola", "universidad"]
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

def generate_squared_tuples(numbers):
    """Genera una lista con tuplas (n, n^2) para una lista de numeros."""
    return [(n, n**2) for n in numbers]

def generate_cubes_greater_than_10(numbers):
    """Genera una lista con cubos de los números mayores a 10."""
    return [n**3 for n in numbers if n > 10]

def generate_even_numbers(numbers):
    """Genera una lista con solo los numeros pares."""
    return [n for n in numbers if n % 2 == 0]

def generate_value_strings(numbers):
    """Dada la lista numeros, producir: ["valor: 10", "valor: 15", ...]."""
    return [f"valor: {n}" for n in numbers]

def convert_celsius_to_fahrenheit(celsius_temps):
    """Genera una lista con la conversion correspondiente a grados Fahrenheit."""
    return [(c * 9/5) + 32 for c in celsius_temps]

def filter_words_by_length(words, min_length=5):
    """Genera una lista filtrando las palabras de mas de 'min_length' letras."""
    return [palabra for palabra in words if len(palabra) > min_length]

def get_initials(words):
    """Genera una lista con solo las iniciales de las palabras."""
    return [palabra[0] for palabra in words]

def classify_numbers_parity(numbers):
    """Genera una lista indicando "par" o "impar" dada una lista de numeros."""
    return ["par" if n % 2 == 0 else "impar" for n in numbers]

def flatten_matrix(matrix):
    """Dada una lista de listas (matriz), aplanala."""
    return [elemento for sublista in matrix for elemento in sublista]

# Ejecución y visualización de resultados

# a. Generar una lista con tuplas (n, n^2) para una lista de numeros.
tuplas_cuadrados = generate_squared_tuples(numeros)
print("a.", tuplas_cuadrados)

# b. Generar una lista con cubos de los números mayores a 10.
cubos_mayores_10 = generate_cubes_greater_than_10(numeros)
print("b.", cubos_mayores_10)

# c. Generar una lista con solo los numeros pares.
numeros_pares = generate_even_numbers(numeros)
print("c.", numeros_pares)

# d. Dada la lista numeros, producir: ["valor: 10", "valor: 15", ...]
lista_valores = generate_value_strings(numeros)
print("d.", lista_valores)

# e. Generar una lista con la conversion correspondiente a grados Fahrenheit.
temperaturas_fahrenheit = convert_celsius_to_fahrenheit(temperaturas_celsius)
print("e.", temperaturas_fahrenheit)

# f. Generar una lista filtrando las palabras de mas de 5 letras.
palabras_mas_de_5 = filter_words_by_length(palabras)
print("f.", palabras_mas_de_5)

# g. Generar una lista con solo las iniciales de las palabras.
iniciales = get_initials(palabras)
print("g.", iniciales)

# h. Generar una lista indicando "par" o "impar" dada una lista de numeros.
pares_e_impares = classify_numbers_parity(numeros)
print("h.", pares_e_impares)

# i. Dada una lista de listas (matriz), aplanala.
lista_aplanada = flatten_matrix(matriz)
print("i.", lista_aplanada)
```
