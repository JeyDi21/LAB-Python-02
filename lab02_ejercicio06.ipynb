```python
# Ejercicio 6 - Minimax (Juego NIM)

# 0. Constantes del juego
INITIAL_STONES = 10
VALID_MOVES = [1, 2, 3]

# 1. Función Minimax para evaluar el estado del juego
def minimax(piedras, es_max):
    # Caso base: si no quedan piedras, el último que movió pierde.
    # Si es el turno de MAX (computadora) y no quedan piedras, significa que el jugador anterior (MIN) ganó.
    # Entonces, MAX pierde (-1). Viceversa para MIN.
    if piedras == 0:
        return -1 if es_max else 1

    # Turno de MAX (Computadora) - Busca maximizar el puntaje
    if es_max:
        mejor_puntaje = -float('inf')
        for movimiento in VALID_MOVES:
            if piedras - movimiento >= 0:
                # Simula el movimiento y llama a minimax para el siguiente turno (MIN)
                puntaje = minimax(piedras - movimiento, False)
                mejor_puntaje = max(mejor_puntaje, puntaje)
        return mejor_puntaje

    # Turno de MIN (Humano) - Busca minimizar el puntaje
    else:
        mejor_puntaje = float('inf')
        for movimiento in VALID_MOVES:
            if piedras - movimiento >= 0:
                # Simula el movimiento y llama a minimax para el siguiente turno (MAX)
                puntaje = minimax(piedras - movimiento, True)
                mejor_puntaje = min(mejor_puntaje, puntaje)
        return mejor_puntaje

# 2. Función para que la CPU encuentre su mejor movimiento
def mejor_movimiento_cpu(piedras):
    mejor_puntaje = -float('inf')
    mejor_movimiento = 0

    # La CPU prueba todas las jugadas posibles (1, 2 o 3)
    for movimiento in VALID_MOVES:
        if piedras - movimiento >= 0:
            # Llama a minimax simulando que el siguiente turno es del Humano (False)
            puntaje = minimax(piedras - movimiento, False)

            # Si encuentra un camino con mejor puntaje, lo elige
            if puntaje > mejor_puntaje:
                mejor_puntaje = puntaje
                mejor_movimiento = movimiento

    return mejor_movimiento

# 3. Simulación del juego
def jugar_nim():
    piedras = INITIAL_STONES
    print("========================================")
    print("        BIENVENIDO AL JUEGO NIM         ")
    print("========================================")
    print(f"Reglas: Hay {INITIAL_STONES} piedras en la pila.")
    print("En cada turno puedes quitar 1, 2 o 3.")
    print("Pierde el jugador que se quede sin piedras para quitar.")
    print("Tú eres MIN (Humano) y yo soy MAX (CPU).\n")

    turno_humano = True # Empieza el humano

    while True:
        print(f"\n--- Quedan {piedras} piedras en la pila ---")

        if piedras <= 0:
            break # El juego ha terminado, se determinará el ganador a continuación

        if turno_humano:
            movimiento = 0
            while True:
                try:
                    movimiento = int(input("Tu turno (MIN). ¿Cuántas piedras quitas? (1, 2 o 3): "))
                    if movimiento not in VALID_MOVES:
                        print(f"Error: Solo puedes quitar {', '.join(map(str, VALID_MOVES))} piedras.")
                    elif piedras - movimiento < 0:
                        print(f"Error: No puedes quitar tantas. Solo quedan {piedras}.")
                    else:
                        break # Movimiento válido
                except ValueError:
                    print("Error: Por favor, ingresa un número entero válido.")

            piedras -= movimiento
            print(f"> Has quitado {movimiento} piedras.")

        else: # Turno de la CPU
            print("Turno de la CPU (MAX)... pensando...")
            movimiento = mejor_movimiento_cpu(piedras)
            piedras -= movimiento
            print(f"> La CPU ha quitado {movimiento} piedras.")

        turno_humano = not turno_humano # Cambiar de turno

    print("\n========================================")
    print("              FIN DEL JUEGO             ")
    print("========================================")

    # El jugador cuyo turno ES cuando piedras <= 0, es quien pierde.
    # Si es turno_humano (True), significa que el humano no puede mover, entonces la CPU ganó.
    # Si no es turno_humano (False), significa que la CPU no puede mover, entonces el humano ganó.
    if turno_humano:
        print("¡Gana la CPU (MAX)! Te has quedado sin movimientos.")
    else:
        print("¡Felicidades! Has ganado (MIN). La CPU se ha quedado sin movimientos.")

jugar_nim()
```
